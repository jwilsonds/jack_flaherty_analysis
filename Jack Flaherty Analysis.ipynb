```python
#list of Flaherty's different pitch types so I can loop through them
pitch_types = ['FF','FT','CH','KC','SL']
```

```python
#this loop calculates how often Flaherty threw each of his pitches in the first half

fh_pitch_freq = []

for i in range(len(pitch_types)):
    
    pitch = first_half.loc[first_half['pitch_type']== pitch_types[i]]
    
    freq = len(pitch)/len(first_half)
    
    fh_pitch_freq.append(freq)
    
```

```python
#this loop calculates how often Flaherty threw each of his pitches in the second half

sh_pitch_freq = []

for i in range(len(pitch_types)):
    
    pitch = second_half.loc[second_half['pitch_type']== pitch_types[i]]
    
    freq = len(pitch)/len(second_half)
    
    sh_pitch_freq.append(freq)
```

```python
#plotting a bar graph that will show the differences in Flaherty's pitch distribution between the 1st and 2nd half

fig, ax = plt.subplots()

ind = np.arange(len(fh_pitch_freq))

width = 0.3

rects1 = ax.bar(ind-width/2, fh_pitch_freq, width, label = 'First half')

rects2 = ax.bar(ind+width/2,sh_pitch_freq, width, label = 'Second Half')

ax.set_ylabel('Pitch Frequency')

ax.set_title('Pitch Frequency Comparison')

ax.set_xticks(ind)

ax.set_xticklabels(('4-Seam','Sinker','Changeup','Knuckle-Curve','Slider'))

ax.legend()
```

```python
#This function spits out a hexbin plot of the strikezone that shows where Flaherty throws the specified pitch most
#valid arguments for pitch are: 'FF','FT','KC','CH','Sl'
def flaherty_pitch_zone(pitch, season_half):
    
    from pybaseball import statcast_pitcher
    
    import matplotlib.pyplot as plt
    
    import pandas as pd
    
    import matplotlib.patches as patches
    
    full_season = statcast_pitcher('2019-03-25','2019-11-30',656427)
    first_half = statcast_pitcher('2019-03-25', '2019-07-09',656427)
    second_half = statcast_pitcher('2019-07-10', '2019-11-30', 656427)
    
    if season_half == '1st':
        
        pitch_df = first_half.loc[first_half['pitch_type']==pitch]
    
    elif season_half == '2nd':
        
        pitch_df = second_half.loc[second_half['pitch_type']==pitch]
        
    else:
        
        output = print('Incorrect argument for season_half. Valid inputs are either 1st or 2nd')
        
    avg = full_season.describe()
    
    #gets the average top measurement of the strikezone for Flaherty 
    top_sz = avg['sz_top'][1]
    
    #gets the average bottom measurement of the strikezone for Flaherty 
    bot_sz = avg['sz_bot'][1]
    
    fig, ax = plt.subplots()
    
    #generates strikezone using the measurements we got above
    strikezone = patches.Rectangle((-0.70833,bot_sz), width= 17/12, height = (top_sz-bot_sz), fill=False)
    
    pitch_df = pitch_df.loc[(pitch_df['plate_z']<= (top_sz+0.75))&(pitch_df['plate_z']>= 0.25)]
    
    output = ax.hexbin(pitch_df['plate_x'],pitch_df['plate_z'], 
                       gridsize=8, mincnt = 5, cmap = 'Reds')
    
    if pitch == 'FF':
        
        text = '4-Seam'
    
    elif pitch == 'FT':
        
        text = 'Sinker'
        
    elif pitch == 'CH':
            
        text = 'Changeup'
        
    elif pitch == 'KC':
        
        text = 'Knuckle-Curve'
        
    elif pitch == 'SL':
        
        text = 'Slider'
        
    else:
        
        output = print('Invalid argument pitch_type. Valid arguments are: FF, FT, CH, KC, SL')
    
    ax.add_patch(strikezone)
    
    ax.axis('equal')
    
    ax.set_title('Jack Flaherty '+season_half+' half '+text+' location')
    
    cbar = fig.colorbar(output)
    
    cbar.set_label('Number of Pitches')
    
    return output
```

```python
#making a dataframe that groups the data by count and pitch type
fh_grouped_counts = first_half.groupby(['count','pitch_type']).size().to_frame()
```

```python
sh_grouped_counts = second_half.groupby(['count','pitch_type']).size().to_frame()
```

```python
#changing column name
fh_grouped_counts = fh_grouped_counts.rename(columns = {0:'amount_thrown'})
sh_grouped_counts = sh_grouped_counts.rename(columns = {0:'amount_thrown'})
```

```python

possible_counts = ['0-0','0-1', '0-2','1-0','2-0','3-0','3-1','3-2','2-1','2-2','1-1','1-2']
```

```python
#looping through the possible counts to calculate Flaherty's throw rate in each count for each of his pitches
fh_throw_rates = {}
for i in range(len(possible_counts)):
    df = fh_grouped_counts.loc[possible_counts[i]]/fh_grouped_counts.loc[possible_counts[i]].sum()
    fh_throw_rates.update({possible_counts[i]:df})
```

```python
fh_throw_rates = pd.concat(fh_throw_rates)
```

```python
#same loop but for the second half data
sh_throw_rates = {}
for i in range(len(possible_counts)):
    df = sh_grouped_counts.loc[possible_counts[i]]/sh_grouped_counts.loc[possible_counts[i]].sum()
    sh_throw_rates.update({possible_counts[i]:df})
```

```python
sh_throw_rates = pd.concat(sh_throw_rates)
```

```python
#calculating the percent change in throw rate between the first and second half
(sh_throw_rates-fh_throw_rates)*100
```
